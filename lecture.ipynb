```python
import pickle#Serialisation du bytes-codes
import pysrt#lecture du fichier srt
import nltk
from nltk.tokenize import word_tokenize
from openpyxl import Workbook


nltk.download('punkt')  #
```

```python
to_remove=['\u200f','\u200e','\u200e\u200f','\u200f\u200e']
```

```python

#Lecture du fichier str ,en utilisant l'enocage utf-8
srt_file = pysrt.open('./Test.srt', encoding='utf-8')
wb = Workbook()
ws = wb.active
#on parcours le ficheirs et on affiche des textes des sous-titres
for subtitle in srt_file:
    text=subtitle.text#texte de sous-titre
    #Interprer les autres cacarteres comme('\n') \u200e left to right,\u200f right to left
    arabic_words = [w.replace('\u200e', '').replace('\u200f', '') for w in word_tokenize(text[0]) if w not in to_remove]
    
    alert_too_long(subtitle.index,arabic_words,subtitle.start.to_time(),subtitle.end.to_time())
    alert_oneWord(subtitle.index,arabic_words,subtitle.start.to_time(),subtitle.end.to_time())
    alert_maybe_numbers_are_missings(subtitle.index,arabic_words,subtitle.start.to_time(),subtitle.end.to_time())
    alert_one_letter(subtitle.index,arabic_words,subtitle.start.to_time(),subtitle.end.to_time())
    alert_three_succ_lettres(subtitle.index,arabic_words,subtitle.start.to_time(),subtitle.end.to_time())
    alert_repetition(subtitle.index,arabic_words,subtitle.start.to_time(),subtitle.end.to_time())
wb.save("alert_text.xlsx")
    

    
    

    

```

```python
def isword_too_long(phrase,word,index,date_debut,date_fin):
    if len(word)>13:
        alert="Word too long"
        text=' '.join(phrase)
        print(alert)
        ecriture_excel(index,text,alert,date_debut,date_fin)
def alert_too_long(index,phrase,date_debut,date_fin):
    for word in phrase:
        isword_too_long(phrase,word,index,date_debut,date_fin)
def alert_oneWord(index,phrase,date_debut,date_fin):
    if len(phrase)==1:
        alert="One word alert"
        text=' '.join(phrase)
        print(alert)
        ecriture_excel(index,text,alert,date_debut,date_fin)
def alert_three_succ_lettres(index,phrase,date_debut,date_fin):
    
    for word in phrase:
        n=len(word)
        if n>3:
            for i in range(n- 2):
                if word[i] == word[i + 1] == word[i + 2]:
                    alert="Three successive letters"
                    text=' '.join(phrase)
                    print(alert)
                    ecriture_excel(index,text,alert,date_debut,date_fin)
def alert_one_letter(index,phrase,date_debut,date_fin):
    for word in phrase:
        if len(word)==1 and word!='و' :
            alert="One letter"
            text=' '.join(phrase)
            print(alert)
            ecriture_excel(index,text,alert,date_debut,date_fin)
def alert_maybe_numbers_are_missings(index,phrase,date_debut,date_fin):
    word_with_numbers=['ريال','رقم','جنيه','عنوان','عدد']
    for word in phrase:
        if word in word_with_numbers:
            alert="Mayber numbers are missings"
            text=' '.join(phrase)
            print(alert)
            ecriture_excel(index,text,alert,date_debut,date_fin)
def alert_repetition(index,phrase,date_debut,date_fin):
    n=len(phrase)
    if n>1:
        for i in range(n-1):
            if phrase[i]==phrase[i+1]:
                alert="répétition des Mot successifs"
                text=' '.join(phrase)
                print(alert)
                ecriture_excel(index,text,alert,date_debut,date_fin)
                return
        alert_repetition_groupe_mots(index,phrase,date_debut,date_fin)

def alert_repetition_groupe_mots(index,phrase,date_debut,date_fin):
    n = len(phrase)
    for taille in range(1, n // 2 + 1):
        for i in range(n - 2 * taille + 1):
            first_sub = phrase[i:i + taille]
            second_sub = phrase[i + taille:i + 2 * taille]            
            if first_sub == second_sub:
                if taille==n/2:
                    alert="Répétition du phrase"
                    text=' '.join(phrase)
                    print(alert)
                    ecriture_excel(index,text,alert,date_debut,date_fin)
                else:
                    alert="Répétition du Mot"
                    text=' '.join(phrase)
                    print(alert)
                    ecriture_excel(index,text,alert,date_debut,date_fin)
def ecriture_excel(index,text,alert,date_debut,date_fin):
    iter=index
    ws[f'A{iter}'] = index
    ws[f'B{iter}'] = text
    ws[f'C{iter}'] = alert
    ws[f'D{iter}'] = date_debut
    ws[f'E{iter}'] = date_fin
    

```

```python
test=["عيب", "فيك", "عيب", "فيك" ,"لاكن"]
alert_repetition_groupe_mots(test,1)
```
